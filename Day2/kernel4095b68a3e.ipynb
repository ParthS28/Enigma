```python
train = pd.read_csv('/kaggle/input/enigma-day2/train.csv')
test = pd.read_csv('/kaggle/input/enigma-day2/test.csv')
```

```python
id = test['Id']
```

```python
train['area'] = train['Length(m)'] * train['Width(m)']
test['area'] = test['Length(m)'] * test['Width(m)']

train['food/month'] = train['Food Availible(Kg)'] / train['Months']
test['food/month'] = test['Food Availible(Kg)'] / test['Months']
# train = train.drop(['Length(m)', 'Width(m)'], axis = 1)
# test = test.drop(['Length(m)', 'Width(m)'], axis = 1)

train = train.drop(['Length(m)', 'Width(m)', 'Months', 'Food Availible(Kg)'], axis = 1)
test = test.drop(['Length(m)', 'Width(m)', 'Months', 'Food Availible(Kg)'], axis = 1)
train
```

```python
types = ['POWs', 'Handicapped', 'Parkki', 'Undesirables', 'Poles', 'Slavs', 'Jews']

slavs_train = train[train['People'] == 'Slavs']
slavs_test = test[test['People'] == 'Slavs']

poles_train = train[train['People'] == 'Poles']
poles_test = test[test['People'] == 'Poles']

pows_train = train[train['People'] == 'POWs']
pows_test = test[test['People'] == 'POWs']

parki_train = train[train['People'] == 'Parkki']
parki_test = test[test['People'] == 'Parkki']


jews_train = train[train['People'] == 'Jews']
jews_test = test[test['People'] == 'Jews']

un_train = train[train['People'] == 'Undesirables']
un_test = test[test['People'] == 'Undesirables']

hands_train = train[train['People'] == 'Handicapped']
hands_test = test[test['People'] == 'Handicapped']
```

```python
slavs_train_x = slavs_train.drop(['Id', 'Survival %', 'People'], axis = 1)
slavs_train_y = slavs_train['Survival %']
slavs_test_x = slavs_test.drop(['Id', 'People'], axis = 1)

poles_train_x = poles_train.drop(['Id', 'Survival %', 'People'], axis = 1)
poles_train_y = poles_train['Survival %']
poles_test_x = poles_test.drop(['Id', 'People'], axis = 1)

pows_train_x = pows_train.drop(['Id', 'Survival %', 'People'], axis = 1)
pows_train_y = pows_train['Survival %']
pows_test_x = pows_test.drop(['Id', 'People'], axis = 1)

parki_train_x = parki_train.drop(['Id', 'Survival %', 'People'], axis = 1)
parki_train_y = parki_train['Survival %']
parki_test_x = parki_test.drop(['Id', 'People'], axis = 1)

jews_train_x = jews_train.drop(['Id', 'Survival %', 'People'], axis = 1)
jews_train_y = jews_train['Survival %']
jews_test_x = jews_test.drop(['Id', 'People'], axis = 1)

un_train_x = un_train.drop(['Id', 'Survival %', 'People'], axis = 1)
un_train_y = un_train['Survival %']
un_test_x = un_test.drop(['Id', 'People'], axis = 1)

hands_train_x = hands_train.drop(['Id', 'Survival %', 'People'], axis = 1)
hands_train_y = hands_train['Survival %']
hands_test_x = hands_test.drop(['Id', 'People'], axis = 1)
```
